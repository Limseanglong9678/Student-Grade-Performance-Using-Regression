```python
import pandas as pd
```

```python
df = pd.read_csv('StudentPerformanceFactors_Cleaned.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = df.select_dtypes(include=["object"]).columns
```

```python
label_encoder = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoder[col] = le
```

```python
df
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
regressor = GradientBoostingRegressor(random_state=0)
regressor.fit(X_train,y_train)
```

```python
first_row_features = X[0].reshape(1, -1)  # Access the first row and reshape it
actual_outcome = y[0]  # Access the first value in the target array
predicted_outcome = regressor.predict(first_row_features)
print("Actual outcome for the first row:", actual_outcome)
print("Predicted outcome for the first row:", predicted_outcome)
```

```python
second_row_features = X[1].reshape(1, -1)  # Access the second row and reshape it
actual_outcome = y[1]  # Access the second value in the target array
predicted_outcome = regressor.predict(second_row_features)
print("Actual outcome for the second row:", actual_outcome)
print("Predicted outcome for the secomd row:", predicted_outcome)
```

```python
third_row_features = X[2].reshape(1, -1)  # Access the third row and reshape it
actual_outcome = y[2]  # Access the third value in the target array
predicted_outcome = regressor.predict(third_row_features)
print("Actual outcome for the third row:", actual_outcome)
print("Predicted outcome for the third row:", predicted_outcome)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = regressor.predict(X_test)
mae = mean_absolute_error(y_test,y_pred)
print("Mean absolue error is",mae)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = regressor.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
print("Mean Squared Error is", mse)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error is ", rmse)
```

```python
from sklearn.metrics import r2_score
y_pred = regressor.predict(X_test)
R_2 = r2_score(y_test,y_pred)
print("R squared score is",R_2)
```
