```python
import pandas as pd
```

```python
df = pd.read_csv('StudentPerformanceFactors_Cleaned.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = df.select_dtypes(include=["object"]).columns
```

```python
label_encoder = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoder[col] = le
```

```python
df
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.reshape(-1, 1))
```

```python
from tensorflow.keras import layers, models
nnr_model = models.Sequential([
    layers.Input(shape=(X_train_scaled.shape[1],)),  # Input layer
    layers.Dense(16, activation='relu'),            # Hidden layer
    layers.Dense(1, activation='linear')            # Output layer (linear activation for regression)
])
```

```python
from tensorflow.keras.optimizers import Adam
# Compile the model
optimizer = Adam(learning_rate=0.001)
nnr_model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])
```

```python
# Train the model
nnr_model.fit(X_train_scaled, y_train_scaled, epochs=100, batch_size=32, validation_data=(X_test_scaled, y_test_scaled), verbose=1)
```

```python
# Predictions
y_pred_scaled = nnr_model.predict(X_test_scaled).flatten()
y_pred = y_scaler.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("Mean absolute error is",mae)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error is", mse)
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error is", rmse)
```

```python
r2 = r2_score(y_test, y_pred)
print("R squared score is",r2)
```
